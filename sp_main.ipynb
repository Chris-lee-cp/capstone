```python
import datetime
```

```python
from sp_linear_reg import SpLinearReg
```

```python
now = datetime.datetime.now()
end_date = now.strftime("%Y-%m-%d")
print end_date

lr_task = SpLinearReg()

lr_task.ivv = lr_task.get_input_feature('IVV', '2016-01-01', end_date)
lr_task.gld = lr_task.get_input_feature('GLD', '2016-01-01', end_date)
#lr_task.ivv = lr_task.get_input_feature_from_file('IVV.csv')
#lr_task.gld = lr_task.get_input_feature_from_file('GLD.csv')

lr_task.preprocess_fetures()

lr_task.display_fetures()

lr_task.do_regression()

lr_task.predict_nextweek()
```

```python
from sp_knn_reg import SpKnnReg
```

```python
now = datetime.datetime.now()
end_date = now.strftime("%Y-%m-%d")
print end_date

kr_task = SpKnnReg()
kr_task.ivv = lr_task.ivv
kr_task.gld = lr_task.gld

kr_task.preprocess_fetures()

kr_task.display_fetures()

kr_task.do_regression()

kr_task.predict_nextweek()

```

```python
from sp_poly_reg import SpPolyReg
```

```python
now = datetime.datetime.now()
end_date = now.strftime("%Y-%m-%d")
print end_date

pr_task = SpPolyReg()
pr_task.ivv = lr_task.ivv
pr_task.gld = lr_task.gld

pr_task.preprocess_fetures()

pr_task.display_fetures()

pr_task.do_regression()

pr_task.predict_nextweek()

```

```python
import matplotlib.pyplot as plt

print "knn regression result"
print kr_task.week_forecast
print "polynomial regression result"
print pr_task.week_forecast

ensemble = (kr_task.week_forecast + pr_task.week_forecast) /2
print "ensemble (knn + polynomial regression) result"
print ensemble

real_prices = lr_task.ivv['Adj Close'][-10:]
ax = real_prices.plot()
ensemble.plot(title='Ensemble stock price forecasting', ax = ax)
plt.show()
```
